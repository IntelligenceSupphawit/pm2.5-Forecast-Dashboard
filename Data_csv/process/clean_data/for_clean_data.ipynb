```python
import pandas as pd

# อ่านไฟล์ CSV
file_path = r"C:\Users\snpdp\pm2.5-Forecast-Dashboard\Data_csv\merge_data\aerosure_wifi_m-1d.csv"
df = pd.read_csv(file_path)

# แปลง timestamp ให้เป็น datetime และเก็บเฉพาะวันที่
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce").dt.date

# ใช้ groupby เพื่อเก็บค่าล่าสุดของแต่ละวัน
df = df.groupby("timestamp").last().reset_index()

# บันทึกทับไฟล์เดิม
df.to_csv(file_path, index=False)

# แสดงผลลัพธ์
print("✅ บันทึกไฟล์เรียบร้อย:", file_path)
```
